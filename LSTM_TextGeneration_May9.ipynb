import keras
keras.__version__

```python
import numpy as np
path=keras.utils.get_file('germanAuthor.txt', origin='https://s3.amazonaws.com/text-datasets/nietzsche.txt')
text=open(path).read().lower()
print('The Corpus Length ->',len(text))
```

```python
#Prepping downloaded corpus data

#this will be the length of partially overlapping character sequence extracted from the text file
maxlen = 10
#this will be the num of char after which a new sequence will be sampled
step=3
#this holds the sentence
sentences = []
#this holds the follow-up characters
next_chars=[]
#sampling logic
for i in range(0,len(text)-maxlen,step):
  sentences.append(text[i:i+maxlen])
  next_chars.append(text[i+maxlen])
print('Number of sentences : ',len(sentences))
#print('next chars -> ',len(next_chars))
#print(range(0,len(text)-maxlen,step))
#range(0,len(text)-maxlen,step)

#listing unique characters in the corpus
chars=sorted(list(set(text)))
#print(chars)
print('Number of unique chars ->',len(chars))

#dictionary mapping of unique chars
chars_indices=dict((char,chars.index(char)) for char in chars)
#print(chars_indices)

#vectorization using one-hot encoding
x=np.zeros((len(sentences),maxlen,len(chars)),dtype=np.bool)
y=np.zeros((len(sentences),len(chars)),dtype=np.bool)
for i, sentence in enumerate(sentences):
  for t, char in enumerate(sentence):
    x[i,t,chars_indices[char]] = 1
  y[i, chars_indices[next_chars[i]]] = 1
```

```python
#net architecture
from keras import layers
model=keras.models.Sequential()
model.add(layers.LSTM(128,input_shape=(maxlen,len(chars))))
model.add(layers.Dense(len(chars),activation='softmax'))
optimiser=keras.optimizers.RMSprop(lr=0.01)
model.compile(loss='categorical_crossentropy',optimizer=optimiser)
```

```python
#training the language model and sampling from it
def sample(preds,temparature=1.0):
     preds = np.asarray(preds).astype('float64')
     preds = np.log(preds) / temparature
     exp_preds=np.exp(preds)
     preds=exp_preds/np.sum(exp_preds)
     probs=np.random.multinomial(1,preds,1)
     return np.argmax(probs)

import random
import sys
for epoch in range(1,60):
  print('epoch',epoch)
  model.fit(x,y,batch_size=128,epochs=1)
  # select a text seed randomly
  start_index=random.randint(0,len(text)-maxlen-1)
  generated_text=text[start_index: start_index+maxlen]
  print('...Generating with seed : "' + generated_text + '" ')
  for temparature in [0.2,0.5,1.0,1.2]:
      print('temparature:', temparature)
      sys.stdout.write(generated_text)

      # generate 400 characters
      for i in range(400):
        sampled = np.zeros((1, maxlen, len(chars)))
        for t, char in enumerate(generated_text):
          sampled[0,t,chars_indices[char]]=1.
          preds = model.predict(sampled, verbose=0)[0]
          next_index = sample(preds, temparature)
          next_char = chars[next_index]

          generated_text += next_char
          generated_text = generated_text[1:]

          sys.stdout.write(next_char)
          sys.stdout.flush()
        print()
```
